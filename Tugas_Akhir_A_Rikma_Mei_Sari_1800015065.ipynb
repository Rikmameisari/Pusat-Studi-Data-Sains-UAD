## Data Preparation

```
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
```

```
#mengimport file csv yang sudah di upload di github
df = pd.read_csv('https://raw.githubusercontent.com/Rikmameisari/Pusat-Studi-Data-Sains-UAD/main/raw_data.csv')
df
```

```
#untuk mengetahui banyaknya data setiap negara yang muncul pada colom location
df.location.value_counts()
```

```
List_location = list(pd.unique(df['location']))
List_location
```

```
#dibentuk data frame baru untuk 'location' (Asia Tenggara) dari data frame df
df_VNM = df[df['location'] == 'Vietnam']
df_IDN = df[df['location'] == 'Indonesia']
df_THA = df[df['location'] == 'Thailand']
df_PHL = df[df['location'] == 'Philippines']
df_MYS = df[df['location'] == 'Malaysia']
df_SGP = df[df['location'] == 'Singapore']
df_LAO = df[df['location'] == 'Laos']
df_MMR= df[df['location'] == 'Myanmar']
df_BRN = df[df['location'] == 'Brunei']
df_TLS = df[df['location'] == 'Timor']
```

```
#menjumlahkan jumlah 'gdp_per_capita' pada setiap 'location' (Asia Tenggara)
sum_VNM = df_VNM.gdp_per_capita.sum()
sum_IDN = df_IDN.gdp_per_capita.sum()
sum_THA = df_THA.gdp_per_capita.sum()
sum_PHL = df_PHL.gdp_per_capita.sum()
sum_MYS = df_MYS.gdp_per_capita.sum()
sum_SGP = df_SGP.gdp_per_capita.sum()
sum_LAO = df_LAO.gdp_per_capita.sum()
sum_MMR = df_MMR.gdp_per_capita.sum()
sum_BRN = df_BRN.gdp_per_capita.sum()
sum_TLS = df_TLS.gdp_per_capita.sum()
```

```
#data untuk negara Indonesia
df_IDN[['location','date', 'total_cases', 'total_deaths','population', 'gdp_per_capita' ]]
```

```
#Melihat pendapatan negara Indonesia selama Pandemic covid-19
List_gdp = list(pd.unique(df_IDN['gdp_per_capita']))
List_gdp
```

```
#dibentuk data frame dari gdp_per_capita pada setiap location (Asia Tenggara)
df_gdp = pd.DataFrame({"location":['Vietnam', 'Indonesia', 'Thailand', 'Philippines', 'Malaysia', 'Singapore', 'Laos', 'Myanmar', 'Brunei', 'Timor'],
              "Pendapatan":[sum_VNM, sum_IDN, sum_THA, sum_PHL, sum_MYS, sum_SGP, sum_LAO, sum_MMR, sum_BRN, sum_TLS]})
df_gdp.sort_values(['Pendapatan', 'location'], ascending=[True, False], inplace=True)
df_gdp
```

## Data Visualization

```
gbr_data = px.bar(df_gdp,
               x = 'location',
               y = 'Pendapatan')

gbr_data.update_layout(
    title = 'Diagram Garis Jumlah pendapatan negara per-capita Selama Pandemic 2019',
    title_font_color = 'green',
    title_font_size = 30,

    xaxis = dict(
        title = 'Negara',
        titlefont_size = 16,
        tickfont_size = 9),
    
    yaxis = dict(
        title = 'jumlah pendapatan',
        titlefont_size = 16,
        tickfont_size = 14))

gbr_data.show()
```
